## Load data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
x_train = pd.read_csv("x_train.csv").values
y_train = pd.read_csv("y_train.csv").values[:,0]
x_test = pd.read_csv("x_test.csv").values
y_test = pd.read_csv("y_test.csv").values[:,0]
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

## 1. Compute the mean vectors mi, (i=1,2) of each 2 classes

```python
## Your code HERE
def compute_mean(x_train, y_train):
    class0 = np.where(y_train == 0)
    class1 = np.where(y_train == 1)
    
    mean_of_class0 = np.mean(x_train[class0], axis = 0)
    mean_of_class1 = np.mean(x_train[class1], axis = 0)
    
    print("mean vectors of class {} is {}".format(0, mean_of_class0))
    print("mean vectors of class {} is {}".format(1, mean_of_class1))
    
    mean = np.vstack((mean_of_class0, mean_of_class1))
    return mean
```

```python
#print(f"mean vector of class 1: {m1}", f"mean vector of class 2: {m2}")
```

## 2. Compute the Within-class scatter matrix SW

```python
## Your code HERE
def compute_within_class(x_train, y_train, mean):
    within_class = np.zeros((2, 2))
    for i in range (750) : 
        within_class += np.dot((x_train[i] - mean[y_train[i]]).reshape(2, 1), (x_train[i] - mean[y_train[i]]).reshape(2, 1).T)
        
    print("within-class scatter matrix is {}".format(within_class))
    
    return within_class
```

```python
#assert sw.shape == (2,2)
#print(f"Within-class scatter matrix SW: {sw}")
```

## 3.  Compute the Between-class scatter matrix SB

```python
## Your code HERE
def compute_between_class(mean):
    between_class = np.dot((mean[0] - mean [1]).reshape(2, 1), (mean[0] - mean [1]).reshape(2, 1).T)
    print("between-class scatter matrix is {}".format(between_class))
    return between_class
```

## 4. Compute the Fisher’s linear discriminant

```python
## Your code HERE
'''def compute_FLD(within_class, between_class) :
    eigen_values, eigen_vectors = np.linalg.eig(np.linalg.inv(within_class).dot(between_class))
    temp = np.array(eigen_values)
    idx = np.where(temp == max(eigen_values))
    discriminant = eigen_vectors[:, idx].reshape(2, 1)
    #print("value is {}, vector is {}".format(eigen_values, eigen_vectors))
    print("fisher's linear discriminant is {}".format(discriminant))'''
def compute_FLD(within_class, mean) :
    discriminant = np.dot(np.linalg.inv(within_class), (mean[1] - mean[0])).reshape(2, 1)
    print("fisher's linear discriminant is {}".format(discriminant))
    
    return discriminant
    
```

## 5. Project the test data by linear discriminant to get the class prediction by nearest-neighbor rule and calculate the accuracy score 
you can use accuracy_score function from sklearn.metric.accuracy_score

```python
#acc = accuracy_score(y_test, y_pred)
def test_prediction(x_train, y_train, project_test, discriminant, temp) :
    project_train = []
    distances = np.zeros(750)
    for i in range(750):
        project_train.append(np.dot(discriminant.T, x_train[i]))
    for i in range(750) : 
        distances[i] = abs(project_test - project_train[i])
    larger = distances.argsort()[:11]
    neighbor = list(y_train[larger])
    y_pred = max(set(neighbor), key=neighbor.count)
    return y_pred
```

## 6. Plot the 1) projection line 2) Decision boundary and colorize the data with each class
### the result should look like this [image](https://i2.kknews.cc/SIG=fe79fb/26q1000on37o7874879n.jpg) (Red line: projection line, Green line: Decision boundary)

```python
def plot(x_train, y_train, discriminant) :
    
    plt.axis('equal')
    plt.xlim(-3, 5)
    plt.ylim(-3, 5)
    
    class0 = x_train[np.where(y_train == 0)]
    class1 = x_train[np.where(y_train == 1)]
    
    plt.scatter(class0[:, 0], class0[:, 1], c = 'b')
    plt.scatter(class1[:, 0], class1[:, 1], c = 'g')
    
    plt.title("Projection line: w = {}, b = {}".format((discriminant[1]/discriminant[0])[0], -1))
    
    project_line = np.poly1d([discriminant[1]/discriminant[0], -1])
    plt.plot([-3, -2, -1, 0, 1, 2, 3], project_line([-3, -2, -1, 0, 1, 2, 3]), c = 'r', lw = 2)
    
    for i in range(len(class0)) : 
        intercept = class0[i][1]+class0[i][0] / (discriminant[1]/discriminant[0]) 
        project_x = (intercept+1) / (discriminant[1]/discriminant[0]+1/(discriminant[1]/discriminant[0]))
        project_y = project_x * (discriminant[1]/discriminant[0]) - 1
        plt.scatter(project_x, project_y, c = 'b')
        plt.plot([class0[i][0], project_x], [class0[i][1], project_y], c = 'gold', lw = 0.3, alpha = 0.5)
        
    for i in range(len(class1)) : 
        intercept = class1[i][1] + class1[i][0] / (discriminant[1]/discriminant[0]) 
        project_x = (intercept+1) / (discriminant[1]/discriminant[0]+1/(discriminant[1]/discriminant[0]))
        project_y = project_x * (discriminant[1]/discriminant[0]) - 1
        plt.scatter(project_x, project_y, c = 'g')
        plt.plot([class1[i][0], project_x], [class1[i][1], project_y], c = 'gold', lw = 0.3, alpha = 0.5)    
```

```python
mean = compute_mean(x_train, y_train)
within_class = compute_within_class(x_train, y_train, mean)
between_class = compute_between_class(mean)
discriminant = compute_FLD(within_class, mean)
y_pred = []
for i in range(250):
    y_pred.append(test_prediction(x_train, y_train, np.dot(discriminant.T, x_test[i]), discriminant, i))
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy of test-set {acc}")
plot(x_train, y_train, discriminant)
```
